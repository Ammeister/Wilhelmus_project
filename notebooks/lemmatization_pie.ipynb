```python
import pandas as pd
from collections import Counter
from glob import glob
import csv
import os.path as path
```

```python
#_____________ PIE-EXTENDED  _________________
```

```python
model_name = "dum"
tagger = get_tagger(model_name, batch_size=256, device="cpu", model_path=None)
```

```python
def lemmatization_pie_extended(path_name, doc_name):
    with open(path_name, encoding="utf8") as file:
        lignes = file.readlines()


        for ligne in lignes:
            iterator, processor = get_iterator_and_processor()
            annotation = tagger.tag_str(ligne, iterator=iterator, processor=processor)
        
    keys = annotation[0].keys()

    with open('csv_textes_annotes_pie/'+doc_name+'.csv', 'w', newline='', encoding="utf-8") as output_file:
        dict_writer = csv.DictWriter(output_file, keys)
        dict_writer.writeheader()
        dict_writer.writerows(annotation)
```

```python
def moulinette(path_name):
    for doc in glob(path_name):
        doc_name = path.splitext(path.basename(doc))[0]
        print(doc_name)
        lemmatization_pie_extended(doc, doc_name)
```

```python
moulinette(path_name)
```

```python
path_name = 'csv_textes_annotes_pie/*.csv'
```

```python
def table_freq(path_name, doc_name):
    csv_courant = pd.read_csv('csv_textes_annotes_pie/'+doc_name+'.csv')
    
    list_lemma = csv_courant['lemma'].values.tolist()
    table_freq_lemma = dict(Counter(list_lemma))
    
    list_pos = csv_courant['pos'].values.tolist()
    table_freq_pos = dict(Counter(list_pos))
    
    
    zipped_lemma = list(zip(table_freq_lemma.keys(), table_freq_lemma.values()))
    df_lemma = pd.DataFrame(zipped_lemma, columns=['Lemma', 'Nombre d\'apparitions'])
    df_lemma['Frequence d\'apparition'] = round(df_lemma['Nombre d\'apparitions']/sum(table_freq_lemma.values()),5)
    
    df_lemma.sort_values('Frequence d\'apparition',ascending=False).to_csv(r'tables_des_frequences_lemma/'+doc_name+'_lemma_freq.csv', index = False)

    
    zipped_pos = list(zip(table_freq_pos.keys(), table_freq_pos.values()))
    df_pos = pd.DataFrame(zipped_pos, columns=['POS', 'Nombre d\'apparitions'])
    df_pos['Frequence d\'apparition'] = round(df_pos['Nombre d\'apparitions']/sum(table_freq_pos.values()),5)
    
    df_pos.sort_values('Frequence d\'apparition',ascending=False).to_csv(r'tables_des_frequences_pos/'+doc_name+'_pos_freq.csv', index = False)
```

```python
def moulinette_table_freq(path_name):
    for doc in glob(path_name):
        doc_name = path.splitext(path.basename(doc))[0]
        print(doc_name)
        table_freq(doc, doc_name)
```

```python
moulinette_table_freq(path_name)
```

```python
path_name = 'txt/*.txt'
```

```python
def three_chars(chaine):
    list_three_chars = []
    for i in range(0, len(chaine)-2):
        list_three_chars.append(chaine[i]+chaine[i+1]+chaine[i+2])   
    return list_three_chars
```

```python
def split_text(path_name, doc_name, n):

    with open(path_name, encoding="utf8") as file:
        lignes = file.readlines()
        for ligne in lignes:
            list_three_chars = three_chars(ligne)
    
    table_freq_chars= dict(Counter(list_three_chars).most_common(n))
    
    zipped_chars = list(zip(table_freq_chars.keys(), table_freq_chars.values()))
    df_chars = pd.DataFrame(zipped_chars, columns=['chars', 'Nombre d\'apparitions'])
    df_chars['Frequence d\'apparition'] = round(df_chars['Nombre d\'apparitions']/sum(table_freq_chars.values()),5)
    
    df_chars.sort_values('Frequence d\'apparition',ascending=False).to_csv(r'three_chars/'+doc_name+'_char_freq.csv', index = False)
```

```python
def moulinette_split(path_name, n):
    for doc in glob(path_name):
        doc_name = path.splitext(path.basename(doc))[0]
        print(doc_name)
        split_text(doc, doc_name, n)
```

```python
moulinette_split(path_name, 100)
```
